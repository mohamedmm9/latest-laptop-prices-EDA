```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import axis as ax
%matplotlib inline
import seaborn as sns
sns.set_style("whitegrid")
```

# reading data

```python
df=pd.read_csv('../input/latest-laptop-price-list/Laptop.csv')
df=df.reset_index(drop=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# visualizing Brands

```python
plt.figure(figsize=(15,8))
plt.xticks(rotation=90)
sns.countplot(x=df['Brand'])
plt.title('number of models for each brand',fontsize=20)
plt.xlabel('Brand',fontsize=15)
plt.ylabel('number of models',fontsize=15)
```

from this plot we can see that DEll his the biggest number of models let's see what is it as percentage

```python
all_models=df['Brand'].value_counts()
all_models_sum=df['Brand'].value_counts().sum()
Dell=all_models[1]
Dell_percent=Dell/all_models_sum * 100
Dell_percent
```

so Dell has about 22.3% from all the models in the market so dell has a high varitey in models

```python
plt.figure(figsize=(15,8))
all_models.plot(kind='pie',fontsize=12)
```

# visualize prices

```python
prices=df['Price']
```

```python
plt.figure(figsize=(15,8))
plt.xticks(rotation=90)
prices.plot(kind='hist')
plt.axvline(prices.mean(), color='r', linestyle='dashed', linewidth=3,label='the mean')
plt.text(prices.mean()*1.025, 25, 'Mean: {:.2f}'.format(prices.mean()),fontsize=20)
```

the red line show the mean value most of prices are average smoe models are so expensive but it is rare

```python
df[['Brand','Model','Price']].groupby('Brand').mean().sort_values(by='Price', ascending = False).head(10).plot(kind = 'barh', grid = True, figsize = (12,8), 
        fontsize = 15, rot =0, legend = None)
plt.axvline(prices.mean(), color='y', linestyle='dashed', linewidth=3,label='the mean')
plt.xlabel('Price', fontsize = 15)
plt.ylabel('Brand', fontsize = 15)
plt.title("prices", fontsize = 15)
plt.show()
```

it is look like apple have the higest models prices Dell have the lowest and compines like acer and HP have average prices
